```python
# Get user transfers: https://fantasy.premierleague.com/drf/entry/33822/transfers
def get_user_transfers(userid):
    url = FPL_URL + '/entry/' + str(userid) + '/transfers'
    r = requests.get(url)
    jsonResponse = r.json()
    seasons = jsonResponse["entry"]

    if not seasons:
        print("no transfers found!")
        return None

    return seasons
```

```python
#get all data 
el_data = [(el['web_name'], el['threat'], el['creativity']) for el in elements if float(el['assists']) > 0 
           or float(el['goals_scored']) > 0]

el_data_clean = [e for e in el_data if float(e[1])>0 and float(e[2])>0]

t = [float(e[1]) for e in el_data_clean]
c = [float(e[2]) for e in el_data_clean]
n = [e[0] for e in el_data_clean]

fig, ax = plt.subplots(figsize=(30,30))
ax.scatter(t, c, s=500, alpha=0.4)
ax.set_xlabel('Threat', fontsize=40)
ax.set_ylabel('Creativity', fontsize=40)



for i, txt in enumerate(n):
    ax.annotate(txt, (t[i], c[i]), fontsize=24, fontname='Arial')
    
```
